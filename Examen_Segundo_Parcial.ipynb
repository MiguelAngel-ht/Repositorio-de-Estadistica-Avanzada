# 9.3.6

Evans et al., examinaron el efecto de la velocidad sobre las fuerzas de reacción del suelo (GRF) en perros con cojera por un ligamento cruzado craneal desgarrado. Los perros se caminaron y trotaron sobre una plataforma de fuerza y se registró el 
GRF (en newtons) durante la fase de apoyo. La siguiente tabla contiene 22 medidas 
de fuerza expresadas como la media de cinco medidas de fuerza por perro al 
caminar y la media de cinco medidas de fuerza por perro al trotar. 

(a) Utilice el valor de GRF al caminar para predecir el valor de GRF al trotar. 

Además, 
para el valor de 𝑋 = 29.4

(b) Construya y grafique el intervalo de confianza al 95 % para 𝜇𝑦|𝑥

(c) Construya y grafique el intervalo de predicción al 95 % para 𝑌.


**Solución**

```r
data <- read.csv('EXR_C09_S03_06.csv',header=TRUE)
#data ---- WALK	 -  TROT
regresion = lm(TROT~WALK, data=data)
summary(regresion)
predict(regresion, data.frame(WALK = 29.4), interval = "confidence")
predict(regresion, data.frame(WALK = 29.4), interval = "prediction")
```

```r
library(tidyverse)
```

```r
confint(regresion, level = 0.95)
ggplot(data, aes(x = WALK, y = TROT)) + geom_point() + geom_smooth(method ="lm", formula = y ~ x, level=0.95)
```

```r
plot(data$WALK,data$TROT)
abline(-1.283, 2.1435)
abline(-13.32,1.75)
abline(-25.16, 1.37)
```

**CONCLUSIÓN** La ecuación obtenida es

$TROT = -13.22 + 1.75\ WALK$

Se rechaza la hipótesis nula, y es posible predecir la fuerza con la que pisan estos perros al trotar conociendo la fuerza con la que caminan.

```r
 X2 = 0; k=1
 for(i in Ex){
   X2 = X2 + (((Ob[k]-i)**2)/i)
   k = k + 1
 }
 "Chi-Squared"
 X2
 "P-value"
 1-pchisq(2.856, 4)
```

Una muestra aleatoria simple de sujetos normales entre las edades de 6 y 18 años arrojó los datos sobre el potasio corporal total (mEq) y el agua corporal total (litros) que se muestran en la siguiente tabla. Deje que el potasio total sea la variable dependiente y utilice la codificación de la variable ficticia para cuantificar la variable cualitativa. Analice los datos mediante técnicas de regresión. Explica los resultados. Trace los datos originales y las ecuaciones de regresión ajustadas.


**Solución**

```r
# CAP 11
data <- read.csv('REV_C11_16.csv',header=TRUE)
#data # POTAS	WATER	SEX
dat <- data.frame(data)
mod_reg <- lm(POTAS~WATER+SEX, data = dat)
summary(mod_reg)
anova(mod_reg)
```

```r
library(dplyr)
library(tidyverse)
```

```r
dat_0 <-filter(dat, SEX=="M")
dat_1 <-filter(dat, SEX=="F")
plot(dat_0$POTAS~dat_0$WATER, main="POTASIO VS AGUA CORPORAL POR GÉNERO",
type = "p", col="blue", xlab = "AGUA CORPORAL", ylab = "NIVEL DE POTASIO")
points(dat_1$WATER, dat_1$POTAS, pch=2, col="red")
legend("topleft",legend=c("VARÓN","MUJER"),
col=c("blue","red"),pch=c(1,2))
```

```r
coefficients(mod_reg)
```

```r
plot(dat_0$POTAS~dat_0$WATER, main="POTASIO VS AGUA CORPORAL POR GÉNERO",
type = "p", col="blue", xlab = "AGUA CORPORAL", ylab = "NIVEL DE POTASIO")
points(dat_1$WATER, dat_1$POTAS, pch=2, col="red")
legend("topleft",legend=c("VARÓN","MUJER"),
col=c("blue","red"),pch=c(1,2))
abline(-34.4208, 79.356) # Method=0 -> HOMBRES
abline(71.1415, 79.356) # Method=1 -> MUJERES
```

**CONCLUSIÓN** El nivel de potasio de un individuo puede ser predecido conociendo el nivel corporal de agua. Hay diferencia entre la predicción para hombres y mujeres, para los hombres el nivel es más alto.
