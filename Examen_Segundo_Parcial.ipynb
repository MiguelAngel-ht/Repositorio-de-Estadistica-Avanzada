```r
dat_0 <-filter(dat, SEX=="M")
dat_1 <-filter(dat, SEX=="F")
plot(dat_0$POTAS~dat_0$WATER, main="POTASIO VS AGUA CORPORAL POR G√âNERO",
type = "p", col="blue", xlab = "AGUA CORPORAL", ylab = "NIVEL DE POTASIO")
points(dat_1$WATER, dat_1$POTAS, pch=2, col="red")
legend("topleft",legend=c("VAR√ìN","MUJER"),
col=c("blue","red"),pch=c(1,2))
```
